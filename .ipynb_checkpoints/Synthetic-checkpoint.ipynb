```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LassoCV
from sklearn.datasets import make_regression
from sklearn.model_selection import KFold

# Example: Generate synthetic regression data
X, y = make_regression(n_samples=100, n_features=20, noise=10, random_state=42)

# LassoCV setup
cv = KFold(n_splits=10, shuffle=True, random_state=42)
lasso_cv = LassoCV(cv=cv, random_state=42).fit(X, y)

# Mean MSE and standard deviation
mse_mean = np.mean(lasso_cv.mse_path_, axis=1)
mse_std = np.std(lasso_cv.mse_path_, axis=1)

# Plotting
plt.figure(figsize=(10, 6))
plt.errorbar(lasso_cv.alphas_, mse_mean, yerr=mse_std, fmt='-o', capsize=3, label='CV MSE ± 1 std. dev')
plt.axvline(lasso_cv.alpha_, linestyle='--', color='red', label=r'Optimal $\lambda^*$')
plt.xscale('log')
plt.xlabel(r'$\lambda$ (log scale)', fontsize=14)
plt.ylabel('Mean Squared Error', fontsize=14)
plt.title('LassoCV: MSE vs. Regularization Parameter', fontsize=16)
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Grid for beta1 and beta2
beta1 = np.linspace(-2, 2, 400)
beta2 = np.linspace(-2, 2, 400)
B1, B2 = np.meshgrid(beta1, beta2)

# Simulated Residual Sum of Squares (RSS) contours (ellipse centered at OLS)
rss = (B1 - 0.5)**2 + 2 * (B2 - 0.5)**2  # Elliptical contours

# Create figure
fig, ax = plt.subplots(figsize=(8, 8))

# Plot RSS contours
contour = ax.contour(B1, B2, rss, levels=15, cmap='gray', linestyles='dotted')
ax.clabel(contour, inline=True, fontsize=8)

# Plot Ridge constraint: L2 ball (circle)
theta = np.linspace(0, 2 * np.pi, 100)
ridge_x = np.cos(theta)
ridge_y = np.sin(theta)
ax.plot(ridge_x, ridge_y, color='blue', linewidth=2, label='Ridge constraint')
ax.fill(ridge_x, ridge_y, color='blue', alpha=0.1)

# Plot Lasso constraint: L1 ball (diamond)
lasso_x = [0, 1, 0, -1, 0]
lasso_y = [1, 0, -1, 0, 1]
ax.plot(lasso_x, lasso_y, color='orange', linewidth=2, label='Lasso constraint')
ax.fill(lasso_x, lasso_y, color='orange', alpha=0.2)

# Mark OLS solution
ax.plot(0.5, 0.5, 'ro', label='OLS solution')
ax.annotate('OLS', xy=(0.5, 0.5), xytext=(0.7, 0.6),
            arrowprops=dict(arrowstyle='->'), fontsize=12)

# Mark Lasso solution (e.g., corner on β1 axis)
ax.plot(1, 0, 'ko')  # sparse solution
ax.annotate('Lasso solution', xy=(1, 0), xytext=(1.2, -0.3),
            arrowprops=dict(arrowstyle='->'), fontsize=12)

# Axes and formatting
ax.axhline(0, color='black', linewidth=0.5)
ax.axvline(0, color='black', linewidth=0.5)
ax.set_xlim(-2, 2)
ax.set_ylim(-2, 2)
ax.set_aspect('equal')

# Labels and legend
ax.set_xlabel(r'$\beta_1$', fontsize=14)
ax.set_ylabel(r'$\beta_2$', fontsize=14)
ax.set_title('RSS Contours with Lasso and Ridge Constraints', fontsize=16)
ax.legend(loc='upper left')
plt.grid(True, linestyle='--', alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from arch import arch_model

# Set seed for reproducibility
np.random.seed(42)

# GARCH(1,1) parameters
n = 1000
omega = 0.1
alpha = 0.1
beta = 0.8

# Define GARCH(1,1) model (zero mean)
am = arch_model(None, mean='Zero', vol='GARCH', p=1, q=1)

# Correct parameter order for simulation: [omega, alpha[1], beta[1]]
params = [omega, alpha, beta]
simulated = am.simulate(params, n)

# Extract returns and conditional variance
returns = simulated['data']
cond_var = simulated['volatility'] ** 2

# Plotting
plt.figure(figsize=(14, 6))

# Plot returns
plt.subplot(2, 1, 1)
plt.plot(returns, label='GARCH(1,1) Returns')
plt.title('GARCH(1,1) Process')
plt.ylabel('Returns')
plt.grid(True)
plt.legend()

# Plot conditional variance
plt.subplot(2, 1, 2)
plt.plot(cond_var, color='orange', label='Conditional Variance')
plt.title('Conditional Variance over Time')
plt.xlabel('Time')
plt.ylabel('Variance')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()

```

```python
# Define the metrics for each window size p
p_values = [5, 10, 20, 40]
mase_values = [0.5821, 0.5914, 0.5974, 0.6032]
r2_values = [0.4357, 0.4334, 0.4289, 0.4289]
fva_values = [41.73, 41.32, 40.97, 40.61]

# Create a 1x3 subplot figure
plt.figure(figsize=(14, 6))



# Plot R²
plt.subplot(1, 3, 1)
plt.plot(p_values, r2_values, marker='o', color='forestgreen', linewidth=3)
plt.title('R² vs. Path Length (p)')
plt.xlabel('Past Window Size (p)')
plt.ylabel('R²')
plt.grid(True)

# Plot MSAE
plt.subplot(1, 3, 2)
plt.plot(p_values, mase_values, marker='o', color='steelblue', linewidth=3)
plt.title('MASE vs. Path Length (p)')
plt.xlabel('Past Window Size (p)')
plt.ylabel('MASE')
plt.grid(True)

# Plot FVA
plt.subplot(1, 3, 3)
plt.plot(p_values, fva_values, marker='o', color='darkred', linewidth=3)
plt.title('FVA (%) vs. Path Length (p)')
plt.xlabel('Past Window Size (p)')
plt.ylabel('FVA (%)')
plt.grid(True)

# Final layout adjustment
plt.tight_layout()
plt.suptitle("Robustness Check: Effect of Past Window Size (p)", fontsize=14, y=1.05)
plt.show()
```
